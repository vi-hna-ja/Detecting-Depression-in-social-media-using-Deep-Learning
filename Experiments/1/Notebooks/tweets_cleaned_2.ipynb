```python
import pandas as pd
```

```python
non_depressed_tweets = "./non_depressed_tweets.csv"
df = pd.read_csv(non_depressed_tweets)
```

```python
def remove_tweets_with_more_hashtags(tweet):
    return True if tweet.count("#") > 5 else False

print("Before removing: ", len(df))
df1 = df
mask1 = df1["tweet"].apply(remove_tweets_with_more_hashtags)
df1 = df1[mask1==False]
print("After removing: ", len(df1))
#print(len(df1))
```

```python
type(df["mentions"][0])
```

```python
mask2a = df1["tweet"].apply(lambda x : len(x) > 25)
df1 = df1[mask2a]
print(len(df1))
```

```python
mask2b = df1["tweet"].apply(lambda x : x.count(" ") > 5)
df1 = df1[mask2b]
print(len(df1))
```

```python
mask3 = df1["urls"].apply(lambda x : len(x) < 5)
df1 = df1[mask3]
print(len(df1))
```

```python
l = df1.urls.tolist()
print(l)
```

```python
import re
df1["mod_text"] = df1["tweet"].apply(lambda x: re.sub(r'#\w+', '', x))
```

```python
df1.mod_text.head()
```

```python
col_list = ["id", "conversation_id", "date", "username", "mod_text", "hashtags", "tweet"]
```

```python
df_final2 = df1[col_list]
df_final2 = df_final2.rename(columns={"mod_text": "tweet_processed", "tweet": "tweet_original"})
```

```python
df_final2["target"] = 0
```

```python
df_final2 = df_final2.reindex(list(range(0, len(df_final2))))
df_final2.to_csv("./non_depressed_tweets_cleaned.csv")
```
